# Zorizko Aldagai Jarraituak

Ordenagailu praktika honetan klasean ikusitako zenbait zorizko jarraituen probabilitateak kalkulatzea dugu helburu.

Zorizko aldagai diskretuetan bezala dentsite-funtzioa (d hasieran erabiliz), banaketa-funtzioa (p hasieran jarriz), alderantzizko kalkulua (q hasieran jarriz) eta zorizko zenbakiak sortzen (r hasieran jarriz)

#### Banaketa normala estandarizatua

Demagun X zorizko aldagai normal estandarizatua dela, N(0,1). Dentsitate-funtzioa kalkulatzeko x puntuan, dnorm komandoa erabiltzen da ondoko eran:

dnorm(x)

Banaketa-funtzioa kalkulatzeko x puntuan, dnorm komandoa erabiltzen da ondoko eran:

pnorm(x)

Demagun p1 zero eta bat arteko balioa dela. Kalkulatzeko x minimoa non P(X<=x) p1 baino handiagoa den, qnorm komandoa erabiltzen da ondoko eran:

qnorm(x)

Zorizko aldagai honi jarraitzen dioten k zenbaki sortzeko rnorm komandoa erabiltzen da ondoko eran:

rnorm(k)

```r
## ADIBIDEA
#  Demagun N(0,1) dugula. Zenbait balio dentsitate-funtzioarenak honakoak dira:
dnorm(0)
dnorm(-3)
dnorm(3)
dnorm(-1)
dnorm(1)
# simetrikoa dela 0-rekiko ikusten da
```

```r
# dentsitate-funtzioaren irudia
x<-seq(-4, 4, by = 0.1)
plot(x,dnorm(x))
```

```r
# banaketa-funtzioaren irudia: pnorm erabiliz, dnorm ordez
x<-seq(-4, 4, by = 0.1)
plot(x,pnorm(x))
```

```r
# Kalkulatzen a non F(a)=0.9
qnorm(0.9)
# Orain F(a)=0.1
qnorm(0.1)
# simetrikoa da!
# Beste balio bat kalkulatuko dugu. Sentzua du?
qnorm(0.5)
```

```r
# 10 balio normal estandarizatuak sorteko 
rnorm(10)
# eta orain beste 10
rnorm(10)
```

### Esponentziala

$\lambda$ parametrodun esponentzialaren probabilitate kalkuluak egiteko dexp(x,$\lambda$), pexp(x,$\lambda$), qexp(x,$\lambda$) eta rexp(k,$\lambda$) erabiltzen dira, aurrekoetan bezala

```r
# Adibidea: dentsitate-funtzioa
lambda<-3
dexp(0.5,3)
# irudia dentsitate-funtzioarena
x<-seq(0, 10, by = 0.1)
plot(x,dexp(x))
```

```r
# irudia banaketa-funtzioarena
plot(x,pexp(x))
```

```r
# Kalkulatzen a non F(a)=0.9
qexp(0.9)
# Orain F(a)=0.2
qexp(0.2)
```

```r
# 10 balio esponentzialak sorteko 
rexp(10)
# eta orain beste 10
rexp(10)
```

### Khi Karratu banaketa

n askatasun graduko Khi karratuko probabilitate kalkuluak egiteko dchisq(x,n), pchisq(x,n), qexp(x,n) eta rexp(k,n) erabiltzen dira, aurrekoetan bezala

```r
## Adibidea: dentsitate-funtzioa: 3 askatasun-graduko KHI karratua
n<-3
dchisq(1,n)
dchisq(5,n)
# dentsitate-funtzioaren irudia
x<-seq(0, 10, by = 0.05)
plot(x,dchisq(x,n))
```

```r
# banaketa-funtzioaren irudia
x<-seq(0, 10, by = 0.05)
plot(x,pchisq(x,n))
```

```r
# Kalkulatzen F(a)=0.9 eta F(a)=0.3
qchisq(0.9,3)
qchisq(0.3,3)
```

```r
# 10 balio banaketa honi jarraitzen diotenak
rchisq(10,3)
# eta beste 10 balio
rchisq(10,3)
```

#### Student-en t- banaketa

n askatasun graduko Student-en t-banaketako probabilitate kalkuluak egiteko dt(x,n), pt(x,n), qt(x,n) eta rt(k,n) erabiltzen dira, aurrekoetan bezala

```r
## Adibidea: dentsitate-funtzioa: 5 askatasun-graduko Student-en t-banaketa
n<-5
dt(0,n)
dt(-1,n)
dt(1,n)
dt(-5,n)
dt(5,n)
# simetrikoa da 0-rekiko! 
# dentsitate-funtzioaren irudia
x<-seq(-5, 5, by = 0.05)
plot(x,dt(x,n))
```

```r
# banaketa-funtzioaren irudia
x<-seq(-5, 5, by = 0.05)
plot(x,pt(x,n))
```

```r
# Kalkulatzen F(a)=0.95 eta F(a)=0.4
qt(0.95,n)
qt(0.4,n)
```

```r
# 10 balio banaketa honi jarraitzen diotenak
rt(10,3)
# eta beste 10 balio
rt(10,3)
```

### Fisher-Snedecor-en Banaketa

n1 eta n2 parametrodun Fisher-Snedecor-en banaketako probabilitate kalkuluak egiteko df(x,n1,n2), pf(x,n1,n2), qf(x,n1,n2) eta rf(k,n1,n2) erabiltzen dira, aurrekoetan bezala

```r
## Adibidea: n1=2 eta n2=5 parametrodun Fisher-Snedecor-en banaketa
n1<-2
n2<-5
df(4,n1,n2)
df(1,n1,n2)
# dentsitate-funtzioaren irudia
x<-seq(0, 7, by = 0.05)
plot(x,df(x,n1,n2))
```

```r
# banaketa-funtzioaren irudia
x<-seq(0, 7, by = 0.05)
plot(x,pf(x,n1,n2))
```

```r
# Kalkulatzen F(a)=0.95 eta F(a)=0.4
qf(0.9,n1,n2)
qf(0.2,n1,n2)
```

```r
# 10 balio banaketa honi jarraitzen diotenak
rf(10,n1,n2)
# eta beste 10 balio
rf(10,n1,n2)
```
