We will see an application of the Central Limit Theorem

## Example: A dice with three outcomes

```r
# we will need to install and load the following package
#install.packages('discreteRV')
library(discreteRV)

# we have seen this command in the computer practise part
X <- RV(outcomes = 1:3, probs = 1/3)
X
plot(X)
```

```r
# the sum of two dice
two_dice<-SofIID(X, n=2)
two_dice
plot(two_dice)
```

```r
# the sum of 3 dice now
hiru_dado<-SofIID(X, n=3)
hiru_dado
plot(hiru_dado)
```

```r
# we now sum 10 dice. We do not show the probabilities, only the illustration
ten_dice<-SofIID(X, n=10)
#ten_dice
plot(ten_dice)
```

```r
# the sum of 20 dice
twenty_dice<-SofIID(X, n=20)
#twenty_dice
plot(twenty_dice)
```

```r
# the sum of 100 dice
houndred_dice<-SofIID(X, n=100)
#houndred_dice
plot(houndred_dice)
```

Suppose that $n$ is big. According to what we have seen above, if we sum the result obtained when we roll n times a dice (with three outcomes), it seems that the resulting follows a normal distribution.

Note that, when we roll a dice with three outcomes, the expected value is equal to 2, in fact
$$
\frac{1}{3}\cdot 1 +\frac{1}{3}\cdot 2 +\frac{1}{3}\cdot 3 = 2
$$

And in all the above cases, the simmetry of the normal distribution is achieved at $2$Eta aurreko kasu guztietan, banaketa normalaren simetria ardatza $n\cdot 2$ dela, hau da, $n$ bider dado baten jaurtiketaren itxaropena.
