### Binomial distribution: probability-law

Suppose that X is a binomial r.v. with parameters n and p, i.e., $X  \sim Bin(n,p)$. To compute the probability-law, we use the command dbinon as follows: 

dbinom(x, n, p)

```r
# EXAMPLE:
# Suppose that we have Bin(10,0.2), that is, n=10 eta p=0.2. 
#   Then, we compute P(X=2) as follows

dbinom(2, 10, 0.2)
```

```r
# We can do a picture as follows
x <- seq(0,10,by = 1)
p_x <- dbinom(x,10,0.2)
plot(x,p_x)
```

### Binomial distribution: distribution function

Suppose that X is a binomial r.v. We compute the distribution at the point x using the command pbinom as follows: 

pbinom(x, n, p)


```r
# EXAMPLE:
# Suppose that we have Bin(10,0.2), that is, n=10 eta p=0.2. 
#   Then, we compute P(X<=2) as follows

pbinom(2, 10, 0.2)
```

```r
# We check that the above result is true by computing this: P(X=0)+P(X=1)+P(X=2)
dbinom(2, 10, 0.2)+dbinom(1, 10, 0.2)+dbinom(0, 10, 0.2)
```

### Binomial distribution: reverse calculation

Suppose that X is a binomial r.v. and p1 is a value between zero and one. We compute the minimum x such that P(X<=x) is larger than p1 using the command qbinom as follows:

qbinom(p1, n, p)


```r
# EXAMPLE:
# Suppose that we have Bin(10,0.2), that is, n=10 eta p=0.2. 
#   Then, we compute P(X<=2) as follows 
#   Orduan, we compute the minimum x such that P(X<=x)>0.9 as follows

qbinom(0.9, 10, 0.2)
```

```r
# We check that the previous result is correct and we compute P(X<=4) and P(X<=3)
pbinom(4, 10, 0.2)
pbinom(3, 10, 0.2)
# We see below that 4 is the minimum value such that its distribution function is larger than 0.9
```

### Binomial distribution: obtaining random numbers

Suppose that X is a binomial r.v. We obtain k random numbers that follow this distribution as follows:

rbinom(k, n, p)

```r
# EXAMPLE
# Suppose that we have Bin(10,0.2) and we obtain 1000 numbers that follow this distribution.

rbinom(1000, 10, 0.2)
```

#### Note: 

Remmember that if Bin(1,p), we have a Bernoulli distribution with parameter p. Therefore, what we saw above can be used to compute probabilities for this case as well. 

### Poisson distribution: probability-law

Suppose that X is a r.v. that follows the Poisson distribution with parameter lambda, Poi(lambda). 
We compute the probability-law at the point x using the command dpois as follows:

dpois(x, n, p)

```r
# EXAMPLE
# Suppose that X:Poi(4). 
#   Then, we compute P(X=2) as follows

dpois(2, 4)
```

```r
# We create an image with the values of the probability-law
x<-seq(0,13)
plot(dpois(x,4))
```

### Poisson distribution: distribution function

Suppose that X is a r.v. that follows the Poisson distribution with parameter lambda, Poi(lambda). We compute the distribution function at the point x using the command ppois as follows:

dpois(x, n, p)

```r
# EXAMPLE
# Suppose that X:Poi(4) 
#   Then, we compute P(X<=2) in the following way

ppois(2, 4)
```

```r
# We check that the above result is correct by computing the following: P(X=0)+P(X=1)+P(X=2)
sum(dpois(seq(0,2), 4))
```

### Poisson distribution: reverse calculus

Suppose that X is a r.v. that follows the Poisson distribution with parameter lambda, Poi(lambda). For a given p1, we compute the minimum value of x such that P(X<=x) is larger than p1 using the command qpois as follows:

qpois(p1, n, p)

```r
# EXAMPLE
# Suppose that we have X:Poi(4). 
#   Then, the minimum x such that P(X<=x) is larger than 0.8 can be computed as follows
qpois(0.8, 4)
```

```r
# We check the previous result is correct
```

```r
ppois(6, 4)
ppois(5, 4)
```

### Poisson distribution: obtaining random numbers

Suppose that X is a r.v. that follows the Poisson distribution with parameter lambda, Poi(lambda). We obtain k numbers that follow this distribution using the command rbinom as follows:

rpois(k, lambda)

```r
# EXAMPLE
# Suppose that we have X:Poi(4) and we pick 1000 numbers that follow this distribution.

rpois(1000, 3)
```

### A fair dice

In the next example, we have P(X=i)=1/6, i=1,2,3,4,5,6 and we plot the obtained probabilities.

```r
X <- RV(outcomes = 1:6, probs = 1/6)
plot(X)
```

### Unfair dice

In the next example, we have P(X=i)=1/10, i=1,2,3,4,5 and P(X=6)=1/2 and we plot the obtained probabilities.

```r
X2 <- RV(outcomes = 1:6, probs = c(1/10,1/10,1/10,1/10,1/10,1/2))
plot(X2)
```
